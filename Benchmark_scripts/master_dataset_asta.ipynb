```python


def merge_with_image_data(df_master, image_metadata_csv):

    image_metadata_csv['PStudyTime'] = image_metadata_csv['StudyTime'].apply(lambda x: f'{int(float(x)):06}')
    
   
    image_metadata_csv['PStudyDateTime'] = pd.to_datetime(
        image_metadata_csv['StudyDate'].astype(str) + ' ' + image_metadata_csv['PStudyTime'].astype(str),
        format="%Y%m%d %H%M%S"
    )

    df_master['intime'] = pd.to_datetime(df_master['intime'])
    df_master['outtime'] = pd.to_datetime(df_master['outtime'])

  
    merged_df_list = []

    # Iterate through each unique subject_id
    for subject_id in image_metadata_csv['subject_id'].unique():
        # Filter both DataFrames by the current subject_id
        image_subject = image_metadata_csv[image_metadata_csv['subject_id'] == subject_id]
        master_subject = df_master[df_master['subject_id'] == subject_id]

        # Perform asof merge for the current subject_id
        merged_subject_df = pd.merge_asof(
            image_subject.sort_values('PStudyDateTime'),
            master_subject.sort_values('intime'),
            left_on='PStudyDateTime',
            right_on='intime',
            direction='backward'
        )

        # Filter rows to ensure PStudyDateTime is within the range [intime, outtime]
        merged_subject_df = merged_subject_df[
            (merged_subject_df['PStudyDateTime'] >= merged_subject_df['intime']) & 
            (merged_subject_df['PStudyDateTime'] <= merged_subject_df['outtime'])
        ]

        
        merged_df_list.append(merged_subject_df)


    merged_df = pd.concat(merged_df_list, ignore_index=True)

    return merged_df
```

```python
# merge with the image data
image_metadata_csv = pd.read_csv('/local/data/jchen/physionet.org/files/extracted_csvs/mimic-cxr-2.0.0-metadata.csv') 
df_master = merge_with_image_data(df_master_asta, image_metadata_csv)
```
